## A1. 답안 작성  
답변:

정의: A/B TEST란 두 가지 이상의 변형(A: 대조군, B: 실험군)을 무작위로 나누어진 사용자에게 노출시켜, 특정 목표(CTR, CVR 등)에 더 효과적인 버전을 통계적으로 입증하는 실험 방법입니다. 인과관계 파악을 위해 주로 사용된다.


---


유의사항: A/B 테스트 진행 시 가장 경계해야 할 것은 '변화에 대한 일시적 반응'입니다. 이러한 반응을 일으킬 수 있는 초두 효과, 신기 효과 등의 심리적 요인들에 유의해야합니다. 초두 효과란, 사용자가 기존 방식(A)에 익숙해져 있어, 더 효율적인 새로운 방식(B)을 불편하게 느끼고 거부하는 현상입니다. 신기 효과란, 반대로, B안이 실제로 더 좋아서가 아니라 단지 '새롭기 때문에' 호기심에 클릭해보는 현상입니다. 이러한 심리적 요인으로 편향 될 수 있습니다. 따라서 사용자가 변화에 적응하고 거품이 빠질 때까지 충분한 기간을 두고 관찰해야 정확한 성과를 측정할 수 있습니다.


---


예시 상황: OTT 앱의 'AI 추천' 버튼 도입
사용자가 시청할 작품 선택에 고민이 있는 상황을 가정해보았습니다.

1. 실험 설계

A안 (기존): 평소대로 인기 순위 영화 나열

B안 (변경): 메인 화면 최상단에 화려한 색상의 'AI 맞춤 추천' 버튼 배치

2. 실험 경과 (초두/신기 효과의 개입)

초기 3일차: B안의 클릭률(CTR)이 A안 대비 200% 폭등했습니다.

해석: 담당자는 AI 기능의 신기 효과의 작용으로 잘못된 해석의 가능성이 있음.

2주 후: B안의 클릭률이 급격히 하락하여 A안과 거의 차이가 없어짐.

실제 원인: 사용자들은 기능이 유용해서가 아니라, 버튼이 눈에 띄고 궁금해서 한 번 눌러봤을 뿐입니다. 호기심이 충족되자 더 이상 반응하지 않았습니다.


결론: 기간을 충분히 두고 '신기 효과', '과두 효과'가 사라진 뒤의 '평형 상태' 데이터를 분석해야 합니다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
p1 (기본 전환율): 기존 UI(A안)에서 사용자가 공모전 상세 페이지로 진입하는 비율입니다. 0.25 (25%)로 가정합니다.

p2 (기대 전환율): 이미지 클릭 기능을 추가한 B안에서 기대하는 목표 전환율입니다. 기존 대비 2%p 상승한 0.27 (27%)로 가정합니다.

MDE (최소 감지 효과): 두 집단 간의 차이를 2%p($0.02$)로 설정했습니다.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.

from scipy import stats
import math

# 1. 설정값 가정
alpha = 0.05      # 유의수준
power = 0.80      # 검정력
p1 = 0.25         # 기존안 전환율
p2 = 0.27         # 실험안 기대 전환율

# 2. Z값 계산
z_alpha_half = stats.norm.ppf(1 - alpha/2) # 1.96
z_beta = stats.norm.ppf(power)             # 0.84

# 3. 수식 적용
numerator = (z_alpha_half + z_beta)**2 * (p1 * (1 - p1) + p2 * (1 - p2))
denominator = (p1 - p2)**2
n = numerator / denominator

# 4. 결과 출력
print(f"가정한 p1: {p1}, p2: {p2}")
print(f"계산된 Z_alpha/2: {z_alpha_half:.2f}")
print(f"계산된 Z_beta: {z_beta:.2f}")
print(f"최종 필요한 최소 샘플 사이즈(n): {math.ceil(n)} 명 (각 그룹당)")

#########
'''
[A2-2 결과 요약 및 해석]

1. 산출 결과:
   - p1=0.25, p2=0.27(MDE 2%p) 가정 시, 각 그룹당 최소 {n}명의 데이터가 필요함.
   - 즉, 총 {n * 2}명의 사용자를 대상으로 실험을 진행해야 통계적 유의성을 확보할 수 있음.

2. 통계적 시사점:
   - MDE(최소 감지 효과) 설정: 2%p의 차이를 발견하기 위해 약 1.5만 명의 표본이 필요함.
   - 샘플 사이즈와 정밀도: 만약 1%p의 더 미세한 차이를 검증하려면 필요한 샘플 수는
     약 4배(약 6만 명 이상)로 급격히 늘어나므로, 비즈니스 가치와 리소스를 고려한 MDE 설정이 중요함.

3. 실험 운영 유의사항:
   - 초두 효과(Primacy Effect) 관리: 실험 초기에 사용자가 새로운 UI(이미지 클릭)에 적응하는
     학습 기간(Learning Curve) 동안의 데이터 편향을 막기 위해 최소 1~2주 이상의 기간 확보 권장.
'''
```

## Q2-3. 실험에 필요한 최소 실험 기간을 구해봅시다.

[실험 가설]  
귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱 검정이고,

이유는 이번 실험 데이터가 '상세 페이지 진입 성공'과 '실패'라는 범주형 변수의 빈도로 구성되어 있기 때문입니다. 두 집단(대조군, 실험군) 간의 전환 비이 통계적으로 유무를 판단하기 위해서는 분할표를 이용한 카이제곱 검정이 가장 적합합니다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 실험 데이터 입력
# [상세진입 성공 수, 상세진입 실패 수]
# 대조군(A): 7,513명 중 25% 성공 (7513 * 0.25 = 1878.25)
# 실험군(B): 7,513명 중 27% 성공 (7513 * 0.27 = 2028.51)

control_results = [1878, 5635]  # 대조군(A)
variant_results = [2029, 5484]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 공모전 이미지 클릭 허용이 상세 페이지 진입률에 유의미한 상승 효과를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 두 그룹 간의 진입률 차이는 통계적으로 유의미하지 않습니다.")


#############
```

**결과해석**

2. 가설 검증: 대립가설(H1)을 채택하여, 이미지 클릭 영역 확장이 유저의 정보 탐색 접근성을 개선했음을 입증함.
