## A1. 답안 작성  
답변:A,B Test는 두가지 버전(A,B)을 무작위 사용자 집단에 노출시켜 더 나은 성과를 내는 버전을 찾아서 개선하는 방법이다.
예를 들면, 쇼핑몰 결제 페이지에서 A안으로는 기존의 버튼을, B안으로는 색감을 바꾼 페이지의 버튼을 노출해서 전환율을 비교하고, 더 높은 쪽을 채택하는 것이다.
A/B TEST시 유의사항으로는, A/B를 랜덤하게 나눠서 사용자 집단이 편향되지 않게 해야하는 것과, 초두효과를 주의해야 한다는 점이 있다.
초두효과는 새 디자인에 대한 초기의 호기심으로 단기적인 성과가 과대평가될 수 있으므로, 일정 기간 후 안정화된 데이터를 사용해야 한다는 것이다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변: p1은 기존 UI인 A안에서의 전환율을, p2는 새롭게 변경한 UI인 B안에서의 전환율을 의미한다. 페이지네이션 개선이 전환율을 높일 것이라는 가설을 검증하려고 하기 때문에, p1은 0.25, p2는 0.3으로 가정하였다.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.
from scipy.stats import norm

# p1,p2
p1 = 0.25
p2 = 0.30

# 유의수준 알파, 검정력
alpha = 0.05
power = 0.8

# Z값 계산
z_alpha = norm.ppf(1 - alpha / 2)
z_beta = norm.ppf(power)

# n 구하기
n = ((z_alpha + z_beta) ** 2 * (p1 * (1 - p1) + p2 * (1 - p2))) / ((p1 - p2) ** 2)

print(n)
#########
```

이유는 비교하는 지표가 상세페이지에 진입한 비율을 계산한 데이터이기 때문이다. 서로 독립인 두 집단 간의 비율 차이가 통계적으로 유의한지를 검정해야 하므로, 카이제곱 검정을 써야 한다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

이유는 비교하는 지표가 페이지를 2페이지 이상 이동했는지, 상세페이지에 진입했는지와 같이 정의된 비율 데이터이기 때문이다. 서로 독립인 두 집단 간의 비율 차이가 통계적으로 유의한지를 검정해야 하므로, 카이제곱 검정을 써야 한다.

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성
import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [600, 648]  # 대조군(A)
variant_results = [700, 548]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("페이지네이션 UI를 하단 숫자형으로 변경하는 이 2페이지 이상 이동 비율에 유의미한 차이를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 페이지네이션 UI 변경의 효과가 통계적으로 증명되지 않았습니다.")










#############
```

**결과해석**

답변: 카이제곱 검정 결과, 카이제곱 통계량은 15.7340, p-value는 0.0000729001로 유의수준 α = 0.05보다 매우 작게 나타났다.
따라서 귀무가설을 기각한다.

이는 페이지네이션 UI를 하단 숫자형으로 변경한 실험군(B)이 기존의 점 형태 페이지네이션을 사용한 대조군(A)보다 2페이지 이상 이동한 비율이 유의미하게 높다는 것을 의미한다.
즉, 하단 숫자형 페이지네이션은 유저의 추가 페이지 탐색에 도움을 주는 긍정적인 효과가 있으며, 본 A/B 테스트에서 설정한 대립가설을 지지하는 결과라고 해석할 수 있다.
